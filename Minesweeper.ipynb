```python
import random


class minesweeper_board(object):
    value = 0
    selected = False
    mine = False

    def __init__(self):
        self.selected = False

    def __str__(self):
        return str(minesweeper_board.value)

    def is_mine(self):
        if minesweeper_board.value == -1:
            return True
        return False


class minesweeper_class(object):
    def __init__(self, minesweeper_board_size , no_of_mines , first_row , first_col):
        self.board = [[minesweeper_board() for i in range(minesweeper_board_size)] for j in range(minesweeper_board_size)]
        self.minesweeper_board_size = minesweeper_board_size
        self.no_of_mines = no_of_mines
        self.identify_spots = minesweeper_board_size * minesweeper_board_size - no_of_mines
        self.first_row = first_row
        self.first_col = first_col
        i = 0
        while i < no_of_mines:
            x = random.randint(0, self.minesweeper_board_size-1)
            y = random.randint(0, self.minesweeper_board_size-1)
            if not self.board[x][y].mine and x != self.first_row or y != self.first_col:
                self.add_mine(x, y)
                i += 1

    def __str__(self):
        string_return = " "
        divider = "\n---"

        for i in range(0, self.minesweeper_board_size):
            string_return += " | " + str(i)
            divider += "----"
        divider += "\n"

        string_return += divider
        for y in range(0, self.minesweeper_board_size):
            string_return += str(y)
            for x in range(0, self.minesweeper_board_size):
                if self.board[x][y].mine and self.board[x][y].selected:
                    string_return += " |" + str(self.board[x][y].value)
                elif self.board[x][y].selected:
                    string_return += " | " + str(self.board[x][y].value)
                else:
                    string_return += " |  "
            string_return += " |"
            string_return += divider
        return string_return

    def add_mine(self, x, y):
        self.board[x][y].value = -1
        self.board[x][y].mine = True
        for i in range(x-1, x+2):
            if i >= 0 and i < self.minesweeper_board_size:
                if y-1 >= 0 and not self.board[i][y-1].mine:
                    self.board[i][y-1].value += 1
                if y+1 < self.minesweeper_board_size and not self.board[i][y+1].mine:
                    self.board[i][y+1].value += 1
        if x-1 >= 0 and not self.board[x-1][y].mine:
            self.board[x-1][y].value += 1
        if x+1 < self.minesweeper_board_size and not self.board[x+1][y].mine:
            self.board[x+1][y].value += 1

    def make_move(self, x, y):
        self.board[x][y].selected = True
        self.identify_spots -= 1
        if self.board[x][y].value == -1:
            return False
        if self.board[x][y].value == 0:
            for i in range(x-1, x+2):
                if i >= 0 and i < self.minesweeper_board_size:
                    if y-1 >= 0 and not self.board[i][y-1].selected:
                        self.make_move(i, y-1)
                    if y+1 < self.minesweeper_board_size and not self.board[i][y+1].selected:
                        self.make_move(i, y+1)
            if x-1 >= 0 and not self.board[x-1][y].selected:
                self.make_move(x-1, y)
            if x+1 < self.minesweeper_board_size and not self.board[x+1][y].selected:
                self.make_move(x+1, y)
            return True
        else:
            return True

    def hit_mine(self, x, y):
        return self.board[x][y].value == -1

    def is_winner(self):
        return self.identify_spots == 0


#play game
def play_minesweeper_game():
    minesweeper_board_size = int(input("Choose the Width of the board: "))
    no_of_mines = int(input("Choose the number of mines: "))
    game_over = False
    winner = False
    print("Select your first row_no: ")
    first_row = int(input())
    print("Select your first col_no: ")
    first_col = int(input())
    board = minesweeper_class(minesweeper_board_size , no_of_mines , first_row , first_col)
    board.make_move(first_col, first_row)
    while not game_over:
        print(board)
        print("Make your move:")
        x = int(input("Enter Row_no: "))
        y = int(input("Enter Col_no: "))
        #board.make_move(first_row, first_col)
        board.make_move(y , x)
        game_over = board.hit_mine(y, x)
        if board.is_winner() and game_over == False:
            game_over = True
            winner = True

    print(board)
    if winner:
        print("Congratulations, You Win!")
    else:
        print("You hit a mine, Game Over!")

play_minesweeper_game()
```
